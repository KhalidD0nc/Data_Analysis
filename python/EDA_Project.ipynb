```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('bad-drivers.csv')
df.columns = ['state', 'fatal_collisions_rate', 'speeding_percent', 'alcohol_percent', 'not_distracted_percent', 'no_prior_accidents' ,'insurance_premium' , 'insurance_losses']
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

**Univariate Analysis**:

```python
sns.histplot(data=df, x='fatal_collisions_rate', bins=10)
```

```python
sns.displot(data=df, x='no_prior_accidents')
```

```python
plt.figure(figsize=(15,20))
plt.pie(df['fatal_collisions_rate'] , labels = df['state'], autopct="%1.2f%%") ;
```

**Bivariate Analysis**

```python
df.head(1)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df, x='speeding_percent', y='alcohol_percent');
plt.yticks(range(16, 44, 3))
plt.show();
```

```python
sns.scatterplot(data=df, x='speeding_percent', y='alcohol_percent');
```

```python
sns.scatterplot(data=df, x='speeding_percent', y='insurance_losses')
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df, x='speeding_percent', y='insurance_losses')
plt.show()
```

**Multivariate Analysis**

```python
sns.pairplot(data=df, vars=['fatal_collisions_rate', 'speeding_percent', 'alcohol_percent',])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[['fatal_collisions_rate', 'speeding_percent', 'alcohol_percent']].corr(), annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(11,10))
sns.scatterplot(data=df, x='fatal_collisions_rate', y='speeding_percent', hue='alcohol_percent');
```
